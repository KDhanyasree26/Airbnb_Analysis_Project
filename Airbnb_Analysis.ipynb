Piechart.

```python
room_type=df['room_type'].value_counts()
plt.figure(figsize=(10,10))
plt.pie(room_type,labels=room_type.index,autopct='%1.1f%%')
plt.title('Room Type')
plt.show()
```

Histogram.

```python
plt.figure(figsize=(10,8))
sns.histplot(df[df['price']<500]['price'],bins=50,kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Number of Listings')
plt.show()
```

Barplot.

```python
top_10_neighbourhoods=df['neighbourhood'].value_counts().head(10)
plt.figure(figsize=(10,8))
sns.barplot(x=top_10_neighbourhoods.index,y=top_10_neighbourhoods.values)
plt.title('Top 10 Neighbourhoods')
plt.xlabel('Neighbourhood')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45,ha='right')
plt.show()
```

Histplot.

```python
plt.figure(figsize=(10,8))
sns.histplot(df['availability_365'],bins=30,kde=True,color='pink')
plt.title('Availability in 365 Days')
plt.xlabel('Availability')
plt.ylabel('Number of Listings')
plt.show()
```
